**FINDING THE GDP GROWTH IN ALL THE COUNTRIES**

```python
final_data=[]
for country in df['Country Name'].unique():
  df_pr=df[df['Country Name']==country].copy()
  data=df_pr.values
  gdp_change=[0]
  for i in range(1,len(data)):
    prev=data[i-1][3]
    curr=data[i][3]
    gdp_change.append(round(((curr-prev)/prev)*100,2))
  df_pr = df_pr.assign(GDP_change = gdp_change)
  final_data.append(df_pr)
```

```python
df_pr
```

```python
df_2=pd.concat(final_data,axis=0)
df_2.head()
```

```python
df_2.info()
```

```python
lst = ['IND','ITA', 'USA', 'CHN']

dfs = []
for i in lst:
    dfs.append(df[df['Country Code'] == i])

    df_pr = pd.concat(dfs, axis = 0)
```

```python
fig = px.line(df_pr, x = 'Year', y = 'Value', title = 'GDP Comparison - ' + '|'.join(lst), color = 'Country Name')

pyo.plot(fig, filename = '|'.join(lst) + '.html')
```

```python
def comare_gdp(lst, isOpen):

    dfs = []
    for i in lst:
        dfs.append(df[df['Country Code'] == i])

        df_pr = pd.concat(dfs, axis = 0)

    fig = px.line(df_pr, x = 'Year', y = 'Value', title = 'GDP Comparison - ' + '|'.join(lst),
                  color = 'Country Name')
    pyo.plot(fig, filename = '|'.join(lst) + '.html', auto_open = isOpen)
```

```python
comare_gdp(['IND','USA'], False)
```
